```python
from tkinter import *
import nltk
from textblob import TextBlob # this is used for the sentiment analysis
from newspaper import Article
  
```

```python
# nltk.download('punkt')
```

```python


def Summarize():

    url=url1.get('1.0', 'end').strip()

    article = Article(url) # created a article uel and it is focues on that article from the url now

    article.download() # this bascially will download the articlew rather than refrencing the url 
    article.parse()

    article.nlp() # this the easiest natural language processing method


    title.config(state='normal', )
    author.config(state='normal')
    publication.config(state='normal')
    Summary.config(state='normal')
    sentiment.config(state = 'normal')
    
    title.delete('1.0', 'end')# we want to first delete everything that is in the field
    title.insert('1.0', article.title) # teh 1.0 means from the begenning of the text field 

    author.delete('1.0', 'end')
    author.insert('1.0', article.authors)

    publication.delete('1.0', 'end')
    publication.insert('1.0', article.publish_date)

    Summary.delete('1.0', 'end')
    Summary.insert('1.0', article.summary)

 #the sentiment analysis part
    analysis = TextBlob(article.text) # here we passed all the text from the article for analysis, but we could have also just passed the summary part
    print(analysis.polarity)
    sentiment.delete('1.0', 'end')
    sentiment.insert('1.0', f'Sentiment: {"positive" if analysis.polarity > 0 else "negative" if analysis.polarity < 0 else "neutral"}')
   
   
    title.config(state='disabled', )
    author.config(state='disabled')
    publication.config(state='disabled')
    Summary.config(state='disabled')
    sentiment.config(state = 'disabled')
    # print(f'Title: {article.title}')
    # print(f'Author: {article.authors}')
    # print(f'Publication Date: {article.publish_date}')
    # print(f'Summary: {article.summary}') 
   



#####################################
####  the GUI part with tkinter  ####
#####################################
root = Tk()
root.geometry('1200x700')
root.title('News Article Summerizer')
root.config(bg='lightslateblue')


tlabel =Label(root, text='Title', font=('ariel', 18,'bold' ), bd=0, bg='lightslateblue', fg='black')
tlabel.pack()

title = Text(root, height=1.5, width=160, bd=0)
title.config(state='disable', bg='#dddddd')
title.pack()


alabel =Label(root, text='Author', font=('ariel', 18, 'bold'), bd=0, bg='lightslateblue', fg='black')
alabel.pack()

author = Text(root, height=1.5, width=160, bd=0)
author.config(state='disable', bg='#dddddd')
author.pack()



plabel =Label(root, text='Publications', font=('ariel', 18, 'bold'), bd=0, bg='lightslateblue', fg='black')
plabel.pack()

publication = Text(root, height=1.5, width=160, bd=0)
publication.config(state='disable', bg='#dddddd')
publication.pack()


slabel =Label(root, text='Summary', font=('ariel', 18, 'bold'), bd=0, bg='lightslateblue', fg='black')
slabel.pack()

Summary = Text(root, height=20, width=160, bd=0)
Summary.config(state='disable', bg='#dddddd')
Summary.pack()


sentlabel =Label(root, text='Publications', font=('ariel', 18, 'bold'), bd=0, bg='lightslateblue', fg='black')
sentlabel.pack()

sentiment = Text(root, height=1.5, width=160, bd=0)
sentiment.config(state='disable', bg='#dddddd')
sentiment.pack()


ulabel =Label(root, text='URL', font=('ariel', 18, 'bold'), bd=0, bg='lightslateblue', fg='black')
ulabel.pack()

url1 = Text(root, height=1.5, width=160, bd=0)
url1.config( bg='#dddddd')
url1.pack()

btn = Button(root, text='Summarize', command= Summarize, bg= 'white')

btn.pack()





root.mainloop()
```
